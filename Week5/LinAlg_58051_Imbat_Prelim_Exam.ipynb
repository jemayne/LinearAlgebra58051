```
import numpy as np
import matplotlib.pyplot as plt
```

```
profit = np.array([1000, 998, 1104, 3124])
reach = np.array([500, 700, 900, 1000])
```

```
def month_profit_trace(profit, reach, make_figs=True): 
    if (profit.shape == (4,)) and (reach.shape == (4,)): 
        e_week1 = np.array((reach[0], profit[0]))     
        e_week2 = np.array((reach[1], profit[1]))
        e_week3 = np.array((reach[2], profit[2]))
        e_week4 = np.array((reach[3], profit[3]))
        
        e_month = e_week1 + e_week2 + e_week3 + e_week4     
        e_month_magnitude = np.sqrt((e_month[0]**2)+(e_month[1]**2))
        alpha = 10**-6                                   
        theta = np.arctan(np.sin(e_month_magnitude)/ np.cos(e_month_magnitude) + alpha)
        theta = np.degrees(theta)   
        print(e_month)
        

        plt.figure(figsize=(16,5))
        plt.title('Bebang\'s Month Post Efficiency')
        plt.xlim(0,1.01*np.sum(reach))
        plt.ylim(-np.sum(np.abs(profit)),np.sum(np.abs(profit)))
        plt.xlabel('FB Post Reach Increment')
        plt.ylabel('Profit')
        n = 2

        week1_2 = np.add(e_week1, e_week2)
        week1_2_3 = np.add(e_week1, np.add(e_week2, e_week3))
        week1_2_3_4 = np.add(e_week4, np.add(e_week3, np.add(e_week2, e_week1)))

        plt.quiver(0,0, e_week1[0], e_week1[1], 
                   angles='xy', scale_units='xy',scale=1, color='yellowgreen', width=0.0030,
                   label='Week 1: {:.2f}'.format(np.linalg.norm(e_week1)))
        plt.quiver(e_week1[0],e_week1[1], e_week2[0], e_week2[1], 
                   angles='xy', scale_units='xy',scale=1, color='springgreen', width=0.0030,
                   label='Week 2: {:.2f}'.format(np.linalg.norm(e_week2)))
        plt.quiver(week1_2[0], week1_2[1], e_week3[0], e_week3[1], 
                   angles='xy', scale_units='xy',scale=1, color='magenta', width=0.0030,
                   label='Week 3: {:.2f}'.format(np.linalg.norm(e_week3)))
        
        plt.quiver(week1_2_3[0], week1_2_3[1], e_week4[0], e_week4[1], 
                   angles='xy', scale_units='xy',scale=1, color='red', width=0.0030,
                   label='Week 4: {:.2f}'.format(np.linalg.norm(e_week4)))
        
        plt.quiver(0,0, e_month[0], e_month[1], 
                   angles='xy', scale_units='xy',scale=1, color='lightgreen', width=0.0030,
                   label='Efficiency: {:.2f} @ {:.2f}'.format(e_month_magnitude, theta))
        
        

        plt.legend(loc='upper left')

        if make_figs:
            plt.savefig(f'LinAlg-Lab2-Bebang Post Eff-{int(e_month_magnitude)}@{int(theta)}.png', dpi=300)

        plt.show()
    
    else:
        print('There is an error. Review and try again.')  

 

month_profit_trace(profit, reach, make_figs=True)
```
