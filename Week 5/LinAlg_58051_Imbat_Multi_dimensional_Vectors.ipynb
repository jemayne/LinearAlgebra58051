$$J = 7x + 5y $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
J = np.array([7,5])

plt.quiver(0,0, J[0], J[1], angles= 'xy', scale_units= 'xy',
           scale=1, color= 'LightPink', label='J')

plt.xlim(0,15)
plt.ylim(0,15)
plt.grid()
plt.legend()
plt.show()
```

$$M = \left\{
    \begin{array}\\
        2x + 4y + 6z \\ 
        3x + 7y + 5z \\
        4x +2y + 8z
    \end{array}
\right. $$

```python
M = np.array([
              [2,4,6],
              [3,7,5],
              [4,2,8]
])
```

```python
figure = plt.figure()
ax = figure.gca(projection='3d')

plt.quiver(0,0,0, M[0,0], M[0,1], M[0,2], color= 'Red')
plt.quiver(0,0,0, M[1,0], M[1,1], M[1,2], color= 'Salmon')
plt.quiver(0,0,0, M[2,0], M[2,1], M[2,2], color= 'DarkOliveGreen')


ax.set_xlim(0,15)
ax.set_ylim(0,15)
ax.set_zlim(0,15)
plt.show()
```
